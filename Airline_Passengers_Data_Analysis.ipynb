```python
import matplotlib.pyplot as mat
import seaborn as sea
import pandas as pd
dfs = sea.load_dataset("flights")
df = sea.load_dataset("flights")
print(df)
```

```python
month_colors = {
    "Jan": 'blue',
    "Feb": 'black',
    "Mar": 'red',
    "Apr": 'purple',
    "May": 'green',
    "Jun": 'yellow',
    "Jul": 'cyan',
    "Aug": 'orange',
    "Sep": 'lime',
    "Oct": 'pink',
    "Nov": 'brown',
    "Dec": 'gray'
}
sea.lineplot(x = "year" , y = "passengers", data = df, hue = "month", palette=[month_colors[month] for month in df['month'].unique()])
```

```python
sea.barplot(x = "year" , y = "passengers", data = df, hue = "month", palette=[month_colors[month] for month in df['month'].unique()])
mat.title("Airline Grouth Rate")
```

```python
dfs['growth_rate'] = dfs['passengers'].pct_change(periods=1) * 100
dfs['growth_rate'] = dfs['growth_rate'].fillna(0) 
print(dfs)
sea.lineplot(x = "year" , y = "growth_rate", data = dfs, hue = "month", palette=[month_colors[month] for month in df['month'].unique()])

```
